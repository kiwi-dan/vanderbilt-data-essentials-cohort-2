# Dataset 3: IRS data

```python
#dataset 3
income_df = np.read_csv('irs_county_2016.csv')
income_df.head(2)
income_df.tail(2)
```

```python
income_df_tn = income_df.query('STATE == "TN"')
income_df_tn.head()
```

```python
income_df_tn.shape
```

```python
#get list of column names from df and make a new df with only the provided columns (given in the inclusive_cols list below)
df_cols = income_df_tn.columns.values.tolist() 
#print(df_cols)

inclusive_cols = ['STATE', 'COUNTYNAME', 'agi_stub', 'N1', 'mars1', 'MARS2', 'MARS4', 'N2', 'NUMDEP', 'ELDERLY', 'A00100', 'N02650', 'A02650', 'N02300', 'A02300']

#method 1: create list of column names to be dropped (see method1 using .drop() below)
cols_to_drop = []
for col in df_cols:
    if col in inclusive_cols:
        cols_to_drop = cols_to_drop
    else :
        cols_to_drop.append(col)

#method 1
income_dropped_cols_method1 = income_df_tn.drop(columns = cols_to_drop) #drop all other columns

#method 2
income_dropped_cols_method2 = income_df_tn.loc[:,inclusive_cols] #select inclusive columns
```

```python
#create new labels given the following labels below
new_col_labels = ['state', 'county', 'income_bucket', 'return_count', 'single_returns', 'joint_returns', 'head_of_house_returns', 'exemptions', 'dependents', 'elderly', 'agi', 'returns_with_total_inc','total_inc_amt', 'returns_with_unemployment', 'unemployment_comp']

#assign new labels
income_dropped_cols_method1.columns = new_col_labels

#check to see if labels updated
income_dropped_cols_method1
```

# Week 2 task: create dictionary with new labels for income brackets

```python
income_dropped_cols_method1['income_bucket'].unique() #check unique values in income bucket

income_dict = {0: 'Total',
               1: 'Under $1', 
               2: 'Between 1 and $10,000', 
               3: 'Between 10,000 and $25,000',
               4: 'Between 25,000 and $50,000', 
               5: 'Between 50,000 and $75,000', 
               6: 'Between 75,000 and $100,000', 
               7: 'Between 100,000 and $200,000', 
               8: '$200,000 or more'}
```

```python
income_bucket_df = income_dropped_cols_method1
income_bucket_df['income_bucket'].replace(income_dict, inplace = True)
#income_bucket_df
```

```python
income_county_agg = income_bucket_df
```

```python
income_county_agg = income_county_agg.groupby('county').sum().reset_index()
income_county_agg
```

# Week 3 task: create new columns and iterate through df with iterows()

```python
#create empty average income list
avg_inc_list = []
for index, row_vals in income_county_agg.iterrows():
    income = income_county_agg.loc[index, 'total_inc_amt'] * 1000
    return_count = income_county_agg.loc[index, 'return_count']
    avg_income_temp = round(income/return_count)
    avg_inc_list.append(avg_income_temp)
    
income_county_agg['avg_income'] = avg_inc_list
```

```python
ha_costs_df_tn2.describe()
mean_ha_costincomeratio = 0.990191 #from describe()

ha_cost_income_ratio_above_mean = ha_costs_df_tn2.cost_income_ratio/mean_ha_costincomeratio
ha_cost_income_ratio_above_mean = ha_cost_income_ratio_above_mean > 1 
ha_costs_df_tn2['ratio_above_mean'] = ha_cost_income_ratio_above_mean.astype(int) #astype allows us to get numerical from boolean
#ha_costs_df_tn2

ha_costs_df_tn2.ratio_above_mean.sum() #find number of counties that are above the mean (result gives 48)
```

```python
ha_costs_df_tn2 = pd.get_dummies(ha_costs_df_tn2, columns = ['urban'], drop_first = True) #this creates the variable 'urban_Urban'
ha_costs_df_tn2.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

X = ha_costs_df_tn2[['urban_Urban']]
y = ha_costs_df_tn2.ratio_above_mean 

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 321)

logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)

y_pred = logistic_model.predict(X_test)
print("Our model predicts an accuracy of:")
print(metrics.accuracy_score(y_test, y_pred))

print("The naive model would predict 48/91 counties for an accuracy of:")
print(48/91)
```
