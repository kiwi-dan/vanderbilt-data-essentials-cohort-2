### There are 95 counties in the cancer data set, but only 91 were in the heart attack data set

```python
#method 1
county_differences_method1 = list(set(cancer_costs_df_tn.county) - set(ha_costs_df_tn.county))
county_differences_method1 #looks like Van Buren, Pickett, Hancock, and Moore County are contained in only 1 of the datasets
#note that this solution will only work when the first list has the extra counties and is probably not a good approach

#method 2
county_differences_method2 = cancer_costs_df_tn[~cancer_costs_df_tn.county.isin(ha_costs_df_tn.county)]
#but, again, this solution only appears to work if 1 list is missing all the values in the other list
#perhaps two for loops is the only good approach? I'll try that below in method 3

#method 3: create a list of counties in each dataset and use the set difference() method
cancer_counties = list(cancer_costs_df_tn.county.unique())
ha_counties = list(ha_costs_df_tn.county.unique())
county_differences_method3 = list(set(cancer_counties).difference(ha_counties))
        
print("method 1: ", county_differences_method1)
print("method 2: ", county_differences_method2.loc[:,['county']])
print("method 3: ", county_differences_method3)

#method 3 appears to be the best approach to this problem and will work for unique values in both datasets
```
